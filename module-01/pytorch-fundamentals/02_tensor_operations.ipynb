```python
# Exercise 2: Normalize a tensor to have zero mean and unit variance
print("\nExercise 2: Normalize tensor")
tensor = torch.randn(100, 10)
mean = torch.mean(tensor)
std = torch.std(tensor)
print(f"Mean before normalize: {mean}")
print(f"Standard deviation before normalize: {std}")

normalized_result = (tensor - mean)/std
normalized_mean = torch.mean(normalized_result)
normalized_std = torch.std(normalized_result)
print(f"Mean after normalize: {normalized_mean}")
print(f"Standard deviation after normalize: {normalized_std}")
```

```python
# Exercise 3: Find the indices of the top 3 values in a tensor
print("\nExercise 3: Find top 3 indices")
tensor = torch.randn(10)
# Find indices of top 3 values
# using topk
top_values, top_indices = torch.topk(tensor, 3)
print(top_indices)

# using argsort
indices = torch.argsort(tensor, descending=True)[:3]
print("Indices (using argsort):", indices)
```

```python
# Exercise 4: Implement softmax function
print("\nExercise 4: Implement softmax")
logits = torch.randn(2, 5)  # Batch of 2, 5 classes
# Implement: softmax(x) = exp(x) / sum(exp(x))
exp = torch.exp(logits)
print(exp)
# dim=1: We sum across the columns (the classes).
# keepdim=True: Keeps the shape as (2, 1) instead of (2), allowing for broadcasting later.
sum_exp = torch.sum(exp, dim=1, keepdim=True)
print(sum_exp)
# PyTorch broadcasts (2, 5) / (2, 1) automatically
softmax = exp / sum_exp
print("\nImplement softmax\n", softmax)
# Built-in function to check
built_softmax = torch.softmax(logits, dim=1)
print("\nBuilt-in Softmax:\n", built_softmax)
```

```python
# Exercise 5: Compute cosine similarity between two vectors
print("\nExercise 5: Cosine similarity")
v1 = torch.randn(100)
v2 = torch.randn(100)
# Compute cosine similarity: (v1 . v2) / (||v1|| * ||v2||)

# dot product (v1 . v2)
dot_product = torch.dot(v1, v2)

# Magnitudes (||v1|| and ||v2||)
norm_v1 = torch.norm(v1)
norm_v2 = torch.norm(v2)

# Compute Cosine Similarity
cosine_sim = dot_product / (norm_v1 * norm_v2)

print("Cosine similarity: ", cosine_sim)

# Built-in function
import torch.nn.functional as F
# dim=0 means we compute similarity along the vector dimension.
cosine_sim_builtin = F.cosine_similarity(v1, v2, dim=0)

print(f"Built-in Cosine Similarity: {cosine_sim_builtin.item():.4f}")
```
